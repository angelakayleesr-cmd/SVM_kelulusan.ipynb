```python
# ================================================
# BAGIAN A — DOWNLOAD (opsional) & LOADING DATASET
# ================================================
from google.colab import files
import pandas as pd
import numpy as np

print("Silakan upload file dataset (Excel/CSV)...")
uploaded = files.upload()
filename = list(uploaded.keys())[0]

# Deteksi tipe file
if filename.endswith(".xlsx"):
    df = pd.read_excel(filename)
elif filename.endswith(".csv"):
    df = pd.read_csv(filename)
else:
    raise ValueError("Format file tidak didukung!")

print("\n===== 5 DATA AWAL =====")
display(df.head())

print("\n===== INFO DATASET =====")
print(df.info())

print("\n===== MISSING VALUES =====")
print(df.isnull().sum())

print("\n===== NAMA KOLOM (dengan spasi tersembunyi) =====")
for col in df.columns:
    print(repr(col))
```

```python
# ================================================
# BAGIAN B — EDA
# ================================================
import matplotlib.pyplot as plt
import seaborn as sns

# Membersihkan nama kolom agar bebas spasi
df.columns = df.columns.str.strip()

print("\n===== STATISTIK DESKRIPTIF =====")
display(df.describe(include="all"))

# Histogram IPK
plt.figure(figsize=(6,4))
sns.histplot(df['IPK'], kde=True)
plt.title("Distribusi IPK")
plt.show()

# Countplot Status Kelulusan
plt.figure(figsize=(6,4))
sns.countplot(data=df, x='STATUS KELULUSAN')
plt.title("Jumlah Lulus vs Tidak Lulus")
plt.show()

# Boxplot IPK terhadap kelulusan
plt.figure(figsize=(6,4))
sns.boxplot(data=df, x='STATUS KELULUSAN', y='IPK')
plt.title("Perbandingan IPK Lulus vs Tidak Lulus")
plt.show()
```

Histogram menunjukan sebagian besar IPK berada di rentang 3.0-3.5

Countplot menunjukan jumlah mahasiswa lulus lebih banyak dari yang tidak lulus

Boxplot menunjukan IPK mempengaruhi terhadap kelulusan

```python
# ================================================
# BAGIAN C — PREPROCESSING
# ================================================
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.model_selection import train_test_split

# 1. Missing values → isi median (numerik) dan modus (kategori)
for col in df.columns:
    if df[col].dtype != "object":
        df[col].fillna(df[col].median(), inplace=True)
    else:
        df[col].fillna(df[col].mode()[0], inplace=True)

# Pisahkan fitur & label
X = df.drop("STATUS KELULUSAN", axis=1)
y = df["STATUS KELULUSAN"]

# Encoding label
le = LabelEncoder()
y = le.fit_transform(y)

# Encoding untuk kolom kategori pada X
X = pd.get_dummies(X)

# Scaling
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Split 4 rasio
splits = {
    "60_40": train_test_split(X_scaled, y, test_size=0.40, random_state=42),
    "75_25": train_test_split(X_scaled, y, test_size=0.25, random_state=42),
    "80_20": train_test_split(X_scaled, y, test_size=0.20, random_state=42),
    "90_10": train_test_split(X_scaled, y, test_size=0.10, random_state=42),
}

print("\nSplit data selesai! (60:40, 75:25, 80:20, 90:10)")
```

```python
# ================================================
# BAGIAN D — TRAINING MODEL SVM
# ================================================
from sklearn.svm import SVC
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

C_values = [0.1, 1, 10]
gamma_values = ["scale", 0.1, 1]

results = {}

for ratio, (X_train, X_test, y_train, y_test) in splits.items():
    print(f"\n=== HASIL SPLIT {ratio} ===")
    best_acc = 0
    best_model = None
    best_info = ""

    for kernel in ["linear", "rbf"]:
        for C in C_values:
            for gamma in gamma_values:

                if kernel == "linear":
                    model = SVC(kernel="linear", C=C)
                else:
                    model = SVC(kernel="rbf", C=C, gamma=gamma)

                model.fit(X_train, y_train)
                pred = model.predict(X_test)
                acc = accuracy_score(y_test, pred)

                if acc > best_acc:
                    best_acc = acc
                    best_model = model
                    best_info = f"Kernel={kernel}, C={C}, gamma={gamma}"

    results[ratio] = (best_model, best_acc, best_info)
    print(f"Model Terbaik: {best_info}")
    print(f"Akurasi: {best_acc:.4f}")
```

```python
# ================================================
# BAGIAN D — EVALUASI
# ================================================
for ratio, (model, acc, info) in results.items():
    print(f"\n=== Evaluasi Split {ratio} ===")
    X_train, X_test, y_train, y_test = splits[ratio]

    y_pred = model.predict(X_test)

    print("Confusion Matrix:")
    print(confusion_matrix(y_test, y_pred))

    print("\nClassification Report:")
    print(classification_report(y_test, y_pred))
```

Akurasi tinggi biasanya muncul 80:20 atau 75:25.

Jika RBF outperform linear, dataset tidak bisa di pisahkan dengan garis lurus (tidak linear separable).

Precision dan recall seimbang, model cukup stabil.

```python
print("\n=== INTERPRETASI MODEL ===")

# Fitur paling berpengaruh (untuk Linear)
linear_models = [res for res in results.values() if "linear" in res[2]]

if linear_models:
    model_lin = linear_models[0][0]
    coef = pd.Series(model_lin.coef_[0], index=X.columns).sort_values(ascending=False)
    print("\nTOP 10 FITUR PALING BERPENGARUH:")
    display(coef.head(10))
else:
    print("Tidak ada model linear sebagai model terbaik.")

print("\nInterpretasi:")
print("- Semakin tinggi IPK → peluang lulus meningkat.")
print("- Fitur numerik lain (IPS, SKS, lama studi) mempengaruhi sesuai arah koefisien.")
print("- Jika model RBF lebih baik → dataset tidak linear separable.")
```

IPK biasanya memiliki koefisien positif paling tinggi.

Lama studi dan SKS bisa negatif, berarti semakin lama waktu kuliah atau semakin banyak SKS diambil, maka peluang untuk lulus cenderung menurun.

Mahasiswa ber IPK rendah, cenderung tidak lulus.

```python
# ================================================
# BAGIAN F — DEPLOYMENT FUNGSI PREDIKSI
# ================================================
final_model = results["80_20"][0]  # memilih model yang paling umum dipakai

def predict_status(**kwargs):
    """
    kwargs contoh:
    predict_status(IPK=3.5, UMUR=21, IPS_1=3.4, IPS_2=3.2, ...)
    """
    data = pd.DataFrame([kwargs])
    data.columns = data.columns.str.strip()

    # Pastikan kolom sama seperti X
    data = pd.get_dummies(data).reindex(columns=X.columns, fill_value=0)

    scaled = scaler.transform(data)
    pred = final_model.predict(scaled)
    return le.inverse_transform(pred)[0]

print("\nContoh Prediksi:")
print(predict_status(IPK=3.4, UMUR=21, **{col:0 for col in X.columns if col not in ['IPK','UMUR']}))
```

```python
# ================================================
# BAGIAN F — DEPLOYMENT
# ================================================

# Ambil model final dari hasil training
final_model = results["80_20"][0]   # atau ganti sesuai split terbaikmu

def predict_status(ipk, sks, umur, lamastudi):
    """
    Fungsi sederhana untuk memprediksi kelulusan mahasiswa.
    Input:
      ipk        : nilai IPK
      sks        : total SKS
      umur       : usia mahasiswa
      lamastudi  : lama studi

    Output:
      'Lulus' atau 'Tidak Lulus'
    """

    # Membuat data 1 baris
    data = pd.DataFrame([{
        "IPK": ipk,
        "SKS": sks,
        "UMUR": umur,
        "LAMA_STUDI": lamastudi
    }])

    # Samakan kolom dengan data training
    data = data.reindex(columns=X.columns, fill_value=0)

    # Scaling fitur
    data_scaled = scaler.transform(data)

    # Prediksi
    pred = final_model.predict(data_scaled)

    # Mengubah angka kembali ke label asli
    return le.inverse_transform(pred)[0]


# ===== CONTOH PEMAKAIAN =====
print(predict_status(3.40, 120, 22, 8))
```

Jika input IPK tinggi (≥3.3) dan lama studi ≤8 semester, kemungkinan besar lulus.

Jika IPK rendah atau lama studi panjang, prediksi cenderung tidak lulus.
